## Hypothesis II: Natural Disasters are directly correlated with the economy and affect the countrys GDP (gross domestic product)

```python
yearly_event_count_summary.set_index('Year', inplace=True)
yearly_damage_summary.set_index('Year', inplace=True)
percentage_change_in_GDP = calculate_percentage_change_in_GDP(GDP_by_state_data)
percentage_change_in_GDP.set_index('Year', inplace=True)
```

To start with, let us first look at the natural disaster events that have taken place in the span of 20 years from 2000 to 2020 including Tornadoes, Wildfires, Hurricanes and significant Earthquakes, Tsunamis and Volcano Eruptions

```python
# Storms: Wildfires, Hurricanes & Tornadoes
yearly_event_count_summary[['Wildfires', 'Tornadoes', 'Hurricanes']].plot(figsize=(15,6), ylabel='Event Count')
plt.xticks(range(2000, 2021, 1))
plt.title('Storms between 2000 and 2020')
plt.show()
```

```python
# Significant Earthquakes, Tsunamis & Volcano Eruptions
yearly_event_count_summary[['Tsunamis', 'Volcanoes', 'Earthquakes']].plot(figsize=(15,6), ylabel='Event Count')
plt.xticks(range(2000, 2021, 1))
plt.title('Significant Earthquakes, Tsunamis & Volcano Eruptions between 2000 and 2020')
plt.show()
```

```python
# All Natural Disasters from 2000 to 2020
all_event_count_summary = yearly_event_count_summary[['Tsunamis', 'Volcanoes', 'Earthquakes', 'Wildfires', 'Tornadoes', 'Hurricanes']]
all_event_count_summary.plot(figsize=(15,6), ylabel='Event Count')
plt.xticks(range(2000, 2021, 1))
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.title('All Natural Disasters from 2000 to 2020')
plt.show()
```

```python
# Damage Incurred due to Natural Disasters throughout the years 
yearly_damage_summary['Total_Damage'].plot(figsize=(15,6), ylabel='$Dollars', title='Damage Incurred in $Dollars')
plt.xticks(range(2000, 2021, 1))
plt.show()
```

```python
# Majorly contributing years to the Damage:
yearly_damage_summary[['Damage_Percentage']].plot.bar(figsize=(15,6), ylabel='Percentage %')
```

#### Impacts on the Country's Economy:

Gross domestic product (GDP) is the monetary value of all finished goods and services made within a country during a specific period. GDP provides an economic snapshot of a country, used to estimate the size of an economy and growth rate.

```python
# Growth of GDP (in $Mil) throughout the years from 2000 to 2020
percentage_change_in_GDP['GDP_for_US'].plot(figsize=(15,6), ylabel='$Million', title='GDP growth in $Mil')
plt.xticks(range(2000, 2021, 1))
plt.show()
```

```python
# Change in GDP percentage throughout the years (from the previous year)
percentage_change_in_GDP['GDP_change_in_percentage'].plot(marker= 'o', figsize=(15,6), ylabel='Percentage %', title='GDP Change in Percentage')
plt.xticks(range(2000, 2021, 1))
plt.show()
```

### Zooming In into the spike of events and analyzing its effect on the country's GDP

Natural Disasters Effect on the country's GDP, damaged production capabilities, are offset by increased investments in assets and increased labor. Overall, these empirical studies suggest that the indirect effects of natural disasters significantly reduce economic growth, especially in low-income countries. <sup>[2]</sup>

Hence, the after effects of catastrophic disasters on the production capabilities should be visible and relationable by the change in GDP.

#### Spike of Events 1: 2003 to 2007

Count of Disaster Events:

```python
# Natural Disasters during the spike i.e. from 2003 to 2007
slice_label_2003_to_2007 = pd.IndexSlice[2003:2007]
all_event_count_summary_2003_to_2007 = yearly_event_count_summary.loc[slice_label_2003_to_2007, :]
all_event_count_summary_2003_to_2007
```

```python
# Natural Disasters during the spike i.e. from 2003 to 2007
all_event_count_summary_2003_to_2007[['Tsunamis', 'Volcanoes', 'Earthquakes', 'Wildfires', 'Tornadoes', 'Hurricanes']].plot(figsize=(15,6), ylabel='Event Count')
plt.xticks(range(2003, 2008, 1))
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.title('Natural Disasters between 2003 and 2007')
plt.show()
```

Effect on the Country's GDP:

```python
# GDP effects during and after the Natural Disasters spike i.e. from 2003 to 2009
slice_label_2003_to_2009 = pd.IndexSlice[2003:2009]
percentage_change_in_GDP_2003_to_2009 = percentage_change_in_GDP.loc[slice_label_2003_to_2009, :]
percentage_change_in_GDP_2003_to_2009
```

```python
# GDP effects during and after the Natural Disasters spike i.e. from 2003 to 2009 (from the previous year)
percentage_change_in_GDP_2003_to_2009['GDP_change_in_percentage'].plot(marker= 'o', figsize=(15,6), ylabel='Percentage %', title='GDP Change in Percentage')
plt.xticks(range(2003, 2010, 1))
plt.show()
```

Identifying the highest contributing Disasters in terms of Damage Incurred in $Mil:

```python
# Damage Incurred due to Natural Disasters between those years 
yearly_damage_summary.loc[slice_label_2003_to_2007, 'Total_Damage'].plot(marker= 'o', figsize=(15,6), ylabel='$Dollars', title='Damage Incurred in $Dollars')
plt.xticks(range(2003, 2008, 1))
plt.show()
```

```python
# Highest Contributors to the Damage Incurred
yearly_damage_summary.loc[slice_label_2003_to_2007, 'Hurricanes':'Volcanoes'].sum().plot.pie(ylabel='Natural Disasters', title='Contributors of the Damage Incurred', figsize=(6,6))
plt.legend(loc='center left', bbox_to_anchor=(1.3, 0.5))
plt.show()
```

Identifying the most affected states along with the highest contributing Disasters in that State:

```python
state_wise_information = retrieve_information_required_by_state(natural_disasters_order, natural_disasters_data, \
                                                                state_codes, 2003, 2009, 'Total_Damage')
top_state_contributors = state_wise_information.nlargest(5, 'Total')
top_state_contributors.reset_index(inplace=True, drop=True)
top_state_contributors
```

```python
# Highest Contributing Disasters to the Damage Incurred in the top 5 states
top_disaster_contributors = top_state_contributors.loc[:, 'Hurricanes':'Volcanoes']
for index, state in top_disaster_contributors.iterrows():
    print(top_state_contributors.loc[index, 'State':'State_Name'])
    state.plot.pie(ylabel='Natural Disasters', title='Contributors of the Damage Incurred', figsize=(6,6))
    plt.legend(loc='center left', bbox_to_anchor=(1.3, 0.5))
    plt.show()
```
