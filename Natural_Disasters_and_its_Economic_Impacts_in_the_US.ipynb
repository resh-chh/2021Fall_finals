```python
# Data retrieval & Calculations:
natural_disasters_order = ['Hurricanes', 'Tornadoes', 'Wildfires', 'Tsunamis', 'Earthquakes', 'Volcanoes']
natural_disasters_data = [hurricanes_data, tornadoes_data, wildfires_data, tsunamis_data, earthquake_data,
                          volcanoes_data]
yearly_damage_summary, yearly_event_count_summary = yearly_event_summary(natural_disasters_order,
                                                                         natural_disasters_data)
yearly_event_count_summary.set_index('Year', inplace=True)
yearly_damage_summary.set_index('Year', inplace=True)
percentage_change_in_GDP = calculate_percentage_change_in_GDP(GDP_by_state_data)
percentage_change_in_GDP.set_index('Year', inplace=True)
```

## Hypothesis I: Natural Disasters are directly correlated with the economy and affect the countryâ€™s GDP (gross domestic product)

To start with, let us first look at the natural disaster events that have taken place in the span of 20 years from 2000 to 2020 including Tornadoes, Wildfires, Hurricanes and significant Earthquakes, Tsunamis and Volcano Eruptions

```python
# Storms: Wildfires, Hurricanes & Tornadoes
yearly_event_count_summary[['Wildfires', 'Tornadoes', 'Hurricanes']].plot(figsize=(15,6), ylabel='Event Count')
plt.xticks(range(2000, 2021, 1))
plt.title('Storms between 2000 and 2020')
plt.show()
```

```python
# Significant Earthquakes, Tsunamis & Volcano Eruptions
yearly_event_count_summary[['Tsunamis', 'Volcanoes', 'Earthquakes']].plot(figsize=(15,6), ylabel='Event Count')
plt.xticks(range(2000, 2021, 1))
plt.title('Significant Earthquakes, Tsunamis & Volcano Eruptions between 2000 and 2020')
plt.show()
```

```python
# All Natural Disasters from 2000 to 2020
all_event_count_summary = yearly_event_count_summary[['Tsunamis', 'Volcanoes', 'Earthquakes', 'Wildfires', 'Tornadoes', 'Hurricanes']]
all_event_count_summary.plot(figsize=(15,6), ylabel='Event Count')
plt.xticks(range(2000, 2021, 1))
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.title('All Natural Disasters from 2000 to 2020')
plt.show()
```

```python
# Damage Incurred due to Natural Disasters throughout the years 
yearly_damage_summary['Total_Damage'].plot(figsize=(15,6), ylabel='$Million', title='Damage Incurred in $Mil')
plt.xticks(range(2000, 2021, 1))
plt.show()
```

```python
# Majorly contributing years to the Damage:
yearly_damage_summary[['Damage_Percentage']].plot.bar(figsize=(15,6), ylabel='Percentage %')
```

```python
# Growth of GDP (in $Mil) throughout the years from 2000 to 2020
percentage_change_in_GDP['GDP_for_US'].plot(figsize=(15,6), ylabel='$Million', title='GDP growth in $Mil')
plt.xticks(range(2000, 2021, 1))
plt.show()
```

```python
# Change in GDP percentage throughout the years (from the previous year)
percentage_change_in_GDP['GDP_change_in_percentage'].plot(figsize=(15,6), ylabel='$Million', title='GDP Change in Percentage')
plt.xticks(range(2000, 2021, 1))
plt.show()
```

### Zooming In into the spike of events and analyzing its effect on the country's GDP

Natural Disasters Effect on the country's GDP, damaged production capabilities, are offset by increased investments in assets and increased labor. Overall, these empirical studies suggest that the indirect effects of natural disasters significantly reduce economic growth, especially in low-income countries. <sup>[2]</sup>
